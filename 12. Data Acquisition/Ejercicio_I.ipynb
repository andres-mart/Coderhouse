# Adquisición de JSON desde la web. Ejercicio propuesto.

```python
# Importamos las librerías necesarias:
import requests
import json
import pandas as pd
from pandas import json_normalize
```

```python
# URL del JSON
url = "https://jsonplaceholder.typicode.com/users"

try:
    # Realizamos la solicitud GET a la URL:
    response = requests.get(url)

    # Comprobamos si la solicitud fue exitosa (código de respuesta 200):
    if response.status_code == 200:
        # Parseamos el JSON:
        data = json.loads(response.text)

        # Imprimimos el JSON:
        print(data)
    else:
        print(
            "Error: No se pudo obtener el JSON. Código de respuesta:",
            response.status_code,
        )

except Exception as e:
    print("Error:", str(e))
```

```python
# Creamos un dataframe con la información contenida en el JSON:
df = pd.DataFrame(data)
df = pd.concat([df.drop(["address"], axis=1), json_normalize(df["address"])], axis=1)
df = pd.concat([df.drop(["company"], axis=1), json_normalize(df["company"])], axis=1)
df
```
