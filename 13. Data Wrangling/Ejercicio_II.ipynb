```python
# Importacion de la libreria
import pandas as pd
import requests
import json
url = 'https://jsonplaceholder.typicode.com/users'
```

```python
# Creacion del objeto request
web = requests.get(url)
json.loads(web.text)
```

```python
# Creacion del Data frame
df = pd.DataFrame.from_dict(json.loads(web.text))

df

```
