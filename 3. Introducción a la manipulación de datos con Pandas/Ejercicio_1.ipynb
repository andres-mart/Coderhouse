## Ejercicio de manipulación de datos con `pandas`: Juegos Olímpicos

```python
import pandas as pd
```

### 1. Construir un objeto de series de tiempo con el índice igual al año de los juegos Olímpicos:

```python
df = pd.read_csv('...\\Summer-Olympic-medals-1976-to-2008.csv', encoding='latin')
df.head()
```

### 2. Separar nombres de apellidos en dos columnas distintas usando la coma como separador:

#### Separamos columna 'Athlete' en 'first_name' y 'last_name':

```python
name = df['Athlete'].str.split(expand=True)
name = name.drop(name.columns[[2,3,4,5]], axis='columns')
name.columns = ['first_name', 'last_name']
name
```

#### Borramos columna 'Athlete':

```python
del(df['Athlete'])
```

#### Reordenamos columnas 'first_name' y 'last_name':

```python
df = pd.concat([df, name], axis=1)
cols = list(df)
cols.insert(4, cols.pop(cols.index('first_name')))
cols.insert(5, cols.pop(cols.index('last_name')))
df = df[cols]

# Controlamos como queda nuestro DataFrame:
df
```

### 3. Obtener medidas resumen del conjunto de datos. ¿Cuál es el país que ganó mas medallas?

```python
medals = df.groupby('Country')['Medal'].size().reset_index(name = 'Total')
max_medals = medals.max()

print('El país que más medallas obtuvo es: \n', max_medals)
```

### 4. Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento

```python
gender = df[df['Gender'] == 'Men']
gender_men = gender.groupby('Year')['Gender'].size().reset_index(name = 'Total Categoría Hombres')
cols = ['Year']
gender_men[cols] = gender_men[cols].astype(int)
gender_men.index = gender_men['Year']
gender_men = gender_men.drop('Year', axis = 'columns')
gender_men
```
