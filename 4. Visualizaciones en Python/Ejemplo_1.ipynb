## Visualizaciones de datos en Python I

Comenzamos por un gr√°ficos simples:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,21)
y= np.arange(101,121)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

Agregamos t√≠tulo y etiquetas:

```python
x = np.arange(1,21)
y= np.arange(101,121)

fig, ax = plt.subplots()

# T√≠tulo y etiquetas:
plt.title('Aprendiendo Matplotlib')
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

ax.plot(x, y)
plt.show()
```

Valores puntuales en Eje X y Eje Y:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,21)
y= np.arange(101,121)

fig, ax = plt.subplots()

plt.title('Aprendiendo Matplotlib')
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

plt.xticks([5,10,15,20])
plt.yticks([100,105,110,115])

ax.plot(x, y)
plt.show()
```

Editamos la l√≠nea resultante:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,21)
y= np.arange(101,121)

plt.figure(figsize=(10,5))

# Editamos los par√°metros de la linea obtenida en el punto anterior:
plt.plot(x, y, color='purple', linewidth=5, linestyle='--', marker='v', markersize=20, markeredgecolor='black')

plt.title('Aprendiendo Matplotlib', fontdict={'fontsize':15})
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

plt.xticks([5,10,15,20])
plt.yticks([100,105,110,115])

ax.plot(x, y)
plt.show()
```

Agregamos una segunda l√≠nea y leyendas:

```python
x = np.arange(1,21)
y= x*5
y2= x*2
plt.figure(figsize=(10,5))

plt.plot(x, y, color='purple', linewidth=5, linestyle='--', marker='v', markersize=20, markeredgecolor='black')
plt.plot(x, y2, color='green')

plt.title('Aprendiendo Matplotlib', fontdict={'fontsize':15})
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

plt.xticks([5,10,15,20])


ax.plot(x, y)
plt.show()
```

## Barcharts

```python
etiquetas = ['A','B','C','D','E']
valores = [1,3,6,4,5]

plt.bar(etiquetas, valores)
plt.show()
```

Editamos el Barchart:

```python
etiquetas = ['A','B','C','D','E']
valores = [1,3,6,4,5]

barra = plt.bar(etiquetas, valores, color='magenta')

barra[0].set_hatch('/')
barra[1].set_hatch('*')
barra[2].set_hatch('-')

plt.show()
```

Agregamos una l√≠nea de valor promedio:

```python
etiquetas = ['A','B','C','D','E']
valores = [1,3,6,4,5]

barra = plt.bar(etiquetas, valores, color='magenta')

barra[0].set_hatch('/')
barra[1].set_hatch('*')
barra[2].set_hatch('-')

promedio = np.mean(valores)
plt.axhline(promedio, color='darkblue')

plt.show()
```

## Scatter Plots:

```python
x= np.arange(1,21)
y= x*2

plt.scatter(x, y)
plt.show()
```

Duplicamos los marcadores del Scatterplot:

```python
x= np.arange(1,8)
y= x*2
s=[20*2**n for n in range(len(x))]

plt.scatter(x, y, s=s)

plt.show()
```
