Creamos un histograma de la columna "Salary" de "df", utilizando la biblioteca Matplotlib (plt).

La función "plt.hist" se utiliza para crear el histograma. Recibe como parámetro la columna "Salary" de "df".

A continuación, agregamos etiquetas al eje y y al eje x utilizando las funciones "plt.ylabel" y "plt.xlabel", respectivamente. La cadena "Densidad" se utiliza para etiquetar el eje y, lo que significa que el histograma representa una densidad de probabilidad. La cadena "Salario" se utiliza para etiquetar el eje x, lo que indica que el histograma representa la distribución de los salarios.

Se utiliza la función "plt.title" para agregar un título a la gráfica, en este caso "Niveles de salario".

Finalmente, graficamos:

```python
plt.hist(df.Salary)

## Legenda en el eje y
plt.ylabel('Densidad')
 
## Legenda en el eje x
plt.xlabel('Salario')
 
## Título de Gráfica
plt.title('Niveles de salario')

plt.show()
```

```python
# Para poder graficar la distribución de contratos por año, cambiamos el tipo de datos de la columna 'DateofHire':

df['DateofHire'] = pd.to_datetime(df['DateofHire'])
```

```python
# 4. Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire):

serie = sns.displot(data=df, x="DateofHire", kind='kde', fill=True)
serie.set(xlabel='Date of Hire', ylabel='Density', title='Distribucion de contratos por año')
serie.fig.set_size_inches(12, 6)
```

Realizamos una gráfica de densidad (KDE) utilizando la biblioteca Seaborn para visualizar la distribución de los valores en una columna específica del DataFrame.

Primero creamos el gráfico de densidad utilizando la función sns.displot() de Seaborn. Los argumentos utilizados son los siguientes:

* data=df: especificamos el DataFrame que contiene los datos que se utilizarán para crear la gráfica.
* x="DateofHire": especificamos la columna que se utilizará para la variable x de la gráfica de densidad.
* kind='kde': especificamos que se desea un gráfico de densidad (KDE).
* fill=True: especifica que se desea que el área bajo la curva de densidad se rellene con color.

Utilizamos el método set() del objeto 'serie' para establecer las etiquetas de los ejes x e y, y el título de la gráfica.

Finalmente utilizamos el atributo 'fig' del objeto 'serie' para obtener la figura generada por Seaborn, y el método set_size_inches() para establecer el tamaño de la figura en 12 pulgadas de ancho por 6 pulgadas de alto.

En resumen, generamos una gráfica de densidad con Seaborn, establecemos las etiquetas del eje x y del eje y, y el título de la gráfica, y establece el tamaño de la figura.
