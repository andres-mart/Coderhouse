# Estadística y Visualización

## Objetivo:

**1.** Repasar conceptos fundamentales de distribuciones de probabilidad y estadística descriptiva <br />

```python
import seaborn as sns
import scipy.stats
import matplotlib.pyplot as plt
import statistics as stats
import numpy as np
import pandas as pd
```

## 1. Estadística Descriptiva

### 1.1. Medidas de tendencia central

```python
iris = sns.load_dataset('iris')
iris
```

```python
# Calulamos:
## Media:
print('La media de "petal_lenght" es: ', np.mean(iris.petal_length))

## Mediana:
print('La mediana de "petal_length" es: ', stats.median(iris.petal_length))

## Moda:
print('La moda de "petal_lenght" es: ', stats.mode(iris.petal_length))
```

### 1.2. Medidas de Dispersión

```python
# Varianza:
var = np.var(iris.petal_length)

# Desviacion estandar:
import math          # Impotamos la libreria 'math' para utilizar funciones matemáticas

desv = math.sqrt(var)
print('La desviación estandar es: ', desv)
```

```python
# Creamos un gráfico para comparar 'petal_lenght' de diferentes especies,
# nos permite resumir mucho de los datos que hemos visto anteriormente 
# mediante fórmulas:

sns.catplot(data=iris, x='petal_length', y='species', kind='box')
```

## 2. Distribuciones de Probabilidad

### 2.1. Uniforme

```python
from scipy.stats import uniform

# Parametros
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)

ax = sns.displot(data_uniform,
                  bins=100,
                  kde=True,
                  color='skyblue'
                  )
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
plt.show()
```

### 2.2. Normal

```python
from scipy.stats import norm

data_normal = norm.rvs(size=10000,loc=0,scale=1)

ax = sns.distplot(data_normal,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
```

### 2.3. Exponencial

```python
from scipy.stats import expon
data_expon = expon.rvs(scale=1,loc=0,size=1000)

ax = sns.distplot(data_expon,
                  kde=True,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

### 2.4. Binomial

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)

ax = sns.distplot(data_binom,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
```

### 2.5. Bernouilli

```python
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.6)

ax= sns.distplot(data_bern,
                 kde=False,
                 color="skyblue",
                 hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Bernoulli Distribution', ylabel='Frequency')
```
